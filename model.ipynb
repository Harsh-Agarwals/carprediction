```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('car data.csv')
df
```

```python
df.info()
```

```python
df.drop('Car_Name', axis=1, inplace=True)
df.info()
```

```python
df['current year'] = 2022
df.head()
```

```python
df['year difference'] = df['current year'] - df.Year
df.head()
```

```python
df.drop(['Year', 'current year'], axis=1, inplace=True)
df.head()
```

```python
print(df['Transmission'].unique())
print(df['Seller_Type'].unique())
print(df['Fuel_Type'].unique())
print(df['Owner'].unique())
```

```python
num_df = pd.get_dummies(df, drop_first=True)
num_df.head()
```

```python
num_df.shape
```

```python
import seaborn as sns
sns.pairplot(num_df)
```

```python
corrmat = num_df.corr()
features = corrmat.index
plt.figure(figsize=(20, 20))
sns.heatmap(num_df[features].corr(), annot=True, cmap='RdYlGn')
```

```python
X = num_df.iloc[:, 1:]
y = num_df.iloc[:, 0]

X.shape, y.shape
```

```python
X.head()
```

```python
from sklearn.ensemble import ExtraTreesRegressor
```

```python
etr = ExtraTreesRegressor()
etr.fit(X, y)
```

```python
etr.feature_importances_
```

```python
X.columns
```

```python
sns.barplot(X.columns, etr.feature_importances_)
plt.xticks(rotation=70);
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=45)
(X_train.shape, y_train.shape), (X_test.shape, y_test.shape)
```

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor()
```

```python
n_estimators = [x for x in np.arange(100, 1201, 100)]
print(n_estimators)

max_features = ['auto', 'sqrt']
print(max_features)

max_depth = [x for x in np.arange(5, 31, 5)]
print(max_depth)

min_samples_split = [2, 5, 10, 15, 100]
print(min_samples_split)

min_samples_leaf = [1, 2, 5, 10]
print(min_samples_leaf)
```

```python
random_grid = {
    'n_estimators': n_estimators,
    'max_features': max_features,
    'max_depth': max_depth,
    'min_samples_split': min_samples_split,
    'min_samples_leaf': min_samples_leaf
}

random_grid
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
rscv = RandomizedSearchCV(estimator=rf, param_distributions=random_grid, random_state=45, cv=5, verbose=2, n_iter=10)

rscv.fit(X_train, y_train)
```

```python
rscv.best_params_
```

```python
rscv.best_score_
```

```python
preds = rscv.predict(X_test)
preds
```

```python
sns.distplot(y_test - preds)
```

```python
plt.scatter(y_test, preds)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

mae = mean_absolute_error(y_test, preds)
mse = mean_squared_error(y_test, preds)
rmse = np.sqrt(mse)

print('MSE', mse, ', MAE', mae, 'RMSE', rmse)
```
